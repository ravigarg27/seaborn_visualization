```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
np.random.seed(sum(map(ord, "aesthetics")))

%matplotlib inline
```

## Letâ€™s define a simple function to plot some offset sine waves,

```python
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
```

## Matplotlib's plot 

```python
sinplot()
```

## Switch to Seaborn defaults

```python
import seaborn as sns
```

```python
sinplot()
```

## Seaborn splits matplotlib parameters into two independent groups. 
- The first group sets the aesthetic style of the plot. To control the style, use the axes_style() and set_style() functions.
- The second scales various elements of the figure so that it can be easily incorporated into different contexts. To scale the plot, use the plotting_context() and set_context() functions.
* In both cases, the first function returns a dictionary of parameters and the second sets the matplotlib defaults.

## Styling figures
- Five main themes : darkgrid (default), whitegrid, dark, white, ticks

```python
sns.set_style("whitegrid")
data = np.random.normal(size=(20, 6)) + np.arange(6) / 2
sns.boxplot(data=data);
```

```python
sns.set_style("dark")
sinplot()
```

```python
sns.set_style("white")
sinplot()
```

```python
sns.set_style("ticks")
sinplot()
```

## Removing spines from white and ticks with despine() 

```python
sns.set_style("white")
sinplot()
sns.despine()
```

```python
sns.set_style("ticks")
sinplot()
sns.despine()
```

## Offsetting with offset parameter and trimming the axis with trim

```python
sns.violinplot(data=data)
sns.despine(offset=10, trim=True)
```

## Controlling which spine to be removed. despine() has four arguments: top, bottom, right, left. By default top and right are True. If true remove that spine

```python
sns.set_style("whitegrid")
sns.boxplot(data=data, palette="deep")
sns.despine(left=True)
```

## Temporarily setting figure style 

```python
with sns.axes_style("darkgrid"):
    plt.subplot(221)
    sinplot()
with sns.axes_style("whitegrid"):
    plt.subplot(222)
    sinplot()
with sns.axes_style("dark"):
    plt.subplot(223)
    sinplot()
with sns.axes_style("white"):
    plt.subplot(224)
    sinplot()
```

## Scaling plot elements
- Four preset contexts, in order of relative size, are paper, notebook, talk, and poster. The notebook style is the default.

```python
with sns.plotting_context("paper"):
    plt.figure(figsize=(10, 8))
    plt.subplot(221)
    sinplot()
with sns.plotting_context("notebook"):
    plt.subplot(222)
    sinplot()
with sns.plotting_context("talk"):
    plt.subplot(223)
    sinplot()
with sns.plotting_context("poster"):
    plt.subplot(224)
    sinplot()
```
